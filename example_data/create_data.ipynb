```python
# extract connection attributes
connections = []
for line in data['connections']:
    if data['connections'][line]['name'] != '':
        data['connections'][line]['id'] = line
        for i in data['connections'][line]['helpful']:
            data['connections'][line][i] = data['connections'][line]['helpful'][i]
        del data['connections'][line]['helpful']
        connections.append(data['connections'][line])
```

```python
# extract connection attributes
names = {}
for line in data['connections']:
    if data['connections'][line]['name'] != '':
        names[line] = data['connections'][line]['name']
        
# add respondent name
names['respondent'] = respondent['name']
```

```python
# extract edges
edges = []
for line in data['network']:
    if data['network'][line] is True:
        edge = line.split(' to ')
        edges.append((names[edge[0]], names[edge[1]]))
```

```python
# write data to CSV
with open('connections.csv', 'w') as f:
    writer = csv.DictWriter(f, connections[0].keys())
    writer.writeheader()
    writer.writerows(connections)
```

```python
# write data to CSV
with open('network.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(["from", "to"])
    writer.writerows(edges)
```
